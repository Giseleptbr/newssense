# NewsSense — EDA das Notícias Coletadas

Exploração inicial do dataset de notícias financeiras coletadas via RSS.
Reexecute esta célula a célula conforme `data/raw/noticias.csv` for crescendo.

```python
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option("display.max_colwidth", 100)

df = pd.read_csv("../data/raw/noticias.csv")
df["data_publicacao"] = pd.to_datetime(df["data_publicacao"], utc=True, errors="coerce")
df["coletado_em"] = pd.to_datetime(df["coletado_em"], errors="coerce")
print(f"Total de noticias: {len(df)}")
df.head()
```

## 1. Volume por fonte e categoria

```python
df["categoria"].value_counts().plot(kind="bar", title="Noticias por categoria")
plt.ylabel("quantidade")
plt.tight_layout()
plt.show()
```

```python
df["fonte"].value_counts()
```

## 2. Distribuicao temporal da coleta

```python
df.set_index("data_publicacao").resample("D").size().plot(
    kind="line", marker="o", title="Noticias publicadas por dia"
)
plt.ylabel("quantidade")
plt.tight_layout()
plt.show()
```

## 3. Tamanho dos textos (titulo e resumo)

```python
df["titulo_len"] = df["titulo"].str.len()
df["resumo_len"] = df["resumo_rss"].str.len()
df[["titulo_len", "resumo_len"]].describe()
```

```python
df["resumo_len"].plot(kind="hist", bins=20, title="Distribuicao do tamanho do resumo (caracteres)")
plt.xlabel("caracteres")
plt.tight_layout()
plt.show()
```

## 4. Checagem de duplicatas e qualidade
Confirma que a deduplicacao por link esta funcionando e identifica resumos vazios/curtos demais para serem uteis.

```python
print("Links duplicados:", df["link"].duplicated().sum())
print("Resumos vazios:", (df["resumo_rss"].str.len() == 0).sum())
print("Resumos curtos (<20 caracteres):", (df["resumo_rss"].str.len() < 20).sum())
```

## 5. Amostras para leitura manual
Visualizacao rapida de titulo + resumo para ter uma ideia qualitativa do conteudo antes da rotulagem.

```python
for _, row in df.sample(min(5, len(df))).iterrows():
    print(f"[{row['categoria']}] {row['titulo']}")
    print(f"  -> {row['resumo_rss'][:200]}")
    print()
```
